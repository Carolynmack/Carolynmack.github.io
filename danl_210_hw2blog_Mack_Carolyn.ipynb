Use a Seaborn to plot the distribution of pos for each of those artists

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# finding the artists with the top10 most tracks
top_artists = spotify['artist_name'].value_counts().head(10).index

# Filter data
top_artist_tracks = spotify[spotify['artist_name'].isin(top_artists)]
#in case there are missing values have to drop them so it will still plot
top_artist_tracks = top_artist_tracks.dropna(subset=['pos'])


# Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=top_artist_tracks, x='artist_name', y='pos')

```

From looking at this distribution of box plots, all of the top 10 artists have similar median pos distributions. With a median pos of around 30. From the boxplot we can see that all top ten artists have numerous high outliers. Drake has the largest max around 175, where Rihanna has the lowest max value near 140. Additionally from the plot it looks like Rihanna and The Weekend have the smallest IQR.
